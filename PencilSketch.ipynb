**Task 4 Image to Pencil Sketch with Python**

**Importing Module**

```python
import cv2 as cv
from google.colab.patches import cv2_imshow
```

**Image Loading**

```python
image=cv.imread('/content/drive/MyDrive/Ironman1.jpeg')
cv2_imshow(image)
```

**Converting colour image to gray scale**

```python
grayFilter=cv.cvtColor(image,cv.COLOR_BGR2GRAY)
cv2_imshow(grayFilter)
```

**Converting GrayFilter to inverted (negative) image**

```python
invert=cv.bitwise_not(grayFilter)
cv2_imshow(invert)
```

**Blur the inverted image**

```python
blur=cv.GaussianBlur(invert,(51,51),0)
cv2_imshow(blur)
```

**Inverting the blur image**

```python
blurInvert=cv.bitwise_not(blur)
cv2_imshow(blurInvert)
```

**Merge the grayFilter and blurInvert to obtain pencil sketch image**

```python
sketch=cv.divide(grayFilter,blurInvert,scale=250.0)
cv2_imshow(sketch)
```
