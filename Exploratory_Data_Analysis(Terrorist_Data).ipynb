# **Importing libraries**

```python
import numpy as np
import pandas as pd
import matplotlib.pyplot as plt
```

# **Loading Data**

```python
df=pd.read_csv('/content/drive/MyDrive/Dataset/globalterrorismdb_0718dist.csv',encoding ="latin1")
df.head()
```

```python
df.describe()
```

```python
df.info()
```

```python
df.shape
```

```python
df.columns
```

```python
for i in df.columns:
  print(i,end=",")
```

```python
df=df[["iyear","imonth","iday","country_txt","region_txt","provstate","city","latitude","longitude","location","summary","attacktype1_txt","targtype1_txt","gname","motive","weaptype1_txt","nkill","nwound","addnotes"]]
df.head()
```

```python
df.rename(columns={"iyear":"Year","imonth":"Month","iday":"Day","country_txt":"Country","region_txt":"Region","provstate":"Province/State","city":"City","latitude":"Latitude","longitude":"Longitude","location":"Location","summary":"Summary","attacktype1_txt":"Attack Type","targtype1_txt":"Target Type","gname":"Group Name","motive":"Motive","weaptype1_txt":"Weapon Type","nkill":"Killed","nwound":"Wounded","addnotes":"Add Notes"
},inplace=True)
df.head()
```

```python
df.info()
```

```python
df.shape
```

```python
df.isnull().sum()
```

```python
df["Killed"]=df["Killed"].fillna(0)
df["Wounded"]=df["Wounded"].fillna(0)
df["Casualty"]=df["Killed"]+df["Wounded"]
```

```python
df.describe()
```

# **Exploring Data**

**Year wise Attack**

```python
attacks=df["Year"].value_counts(dropna=False).sort_index().to_frame().reset_index().rename(columns={"index":"Year","Year":"Attacks"}).set_index("Year")
attacks.head()
```

```python
attacks.plot(kind="bar",color="skyblue",figsize=(15,6),fontsize=12)
plt.title("Timeline of Attacks",fontsize=15)
plt.xlabel("Years",fontsize=15)
plt.ylabel("No. of Attacks",fontsize=15)
plt.show()
```

```python
yc=df[["Year","Casualty"]].groupby("Year").sum()
yc.head()
```

```python
yc.plot(kind="bar",color="violet",figsize=(15,6))
plt.title("Year wise Casualty",fontsize=15)
plt.xlabel("Years",fontsize=15)
plt.xticks(fontsize=12)
plt.ylabel("No. of Casualty",fontsize=15)
plt.show()
```

```python
yk=df[["Year","Killed"]].groupby("Year").sum()
yk.head()
```

```python
yw=df[["Year","Wounded"]].groupby("Year").sum()
yw.head()
```

```python
fig=plt.figure()
ax0=fig.add_subplot(2,1,1)
ax1=fig.add_subplot(2,1,2)

yk.plot(kind="bar",color="cornflowerblue",figsize=(15,15),ax=ax0)
ax0.set_title("People Killed in each Year")
ax0.set_xlabel("Years")
ax0.set_ylabel("No. of People Killed")

yw.plot(kind="bar",color="cornflowerblue",figsize=(15,15),ax=ax1)
ax1.set_title("People Wounded in each Year")
ax1.set_xlabel("Years")
ax1.set_ylabel("No. of People Wounded")

plt.show()
```

1. Highest no. of Attack held in 2014
2. Most no. of people killed in 2014
3. Most no. of people wounded in 2015 

**Region wise Attack**

```python
reg=pd.crosstab(df.Year,df.Region)
reg.head()
```

```python
reg.plot(kind="area",stacked=False,alpha=0.5,figsize=(20,10))
plt.title("Region wise attacks",fontsize=20)
plt.xlabel("Years",fontsize=20)
plt.ylabel("No. of Attacks",fontsize=20)
plt.show()
```

```python
regt=reg.transpose()
regt["Total"]=regt.sum(axis=1)
ra=regt["Total"].sort_values(ascending=False)
ra
```

```python
ra.plot(kind="bar",figsize=(15,6))
plt.title("Total No. of Attacks in each Region from 1970-2017")
plt.xlabel("Region")
plt.ylabel("Number of Attacks")
plt.show()
```

```python
rc=df[["Region","Casualty"]].groupby("Region").sum().sort_values(by="Casualty",ascending=False)
rc
```

```python
rc.plot(kind="bar",color="cornflowerblue",figsize=(15,6))
plt.title("Region Wise Casualty",fontsize=13)
plt.xlabel("Region",fontsize=13)
plt.xticks(fontsize=11)
plt.ylabel("Number of Casualty",fontsize=13)
plt.show()
```

```python
rk=df[["Region","Killed"]].groupby("Region").sum().sort_values(by="Killed",ascending=False)
rk
```

```python
rw=df[["Region","Wounded"]].groupby("Region").sum().sort_values(by="Wounded",ascending=False)
rw
```

```python
fig=plt.figure()
ax0=fig.add_subplot(1,2,1)
ax1=fig.add_subplot(1,2,2)

rk.plot(kind="bar",color="cornflowerblue",figsize=(15,15),ax=ax0)
ax0.set_title("People Killed in each Region")
ax0.set_xlabel("Regions")
ax0.set_ylabel("No. of People Killed")

rw.plot(kind="bar",color="cornflowerblue",figsize=(15,15),ax=ax1)
ax1.set_title("People Wounded in each Region")
ax1.set_xlabel("Regions")
ax1.set_ylabel("No. of People Wounded")

plt.show()
```

1. No. of terrorist activities are high in Middle East & North Africa
2. Most no. of casualty was also in Middle East & North Africa
